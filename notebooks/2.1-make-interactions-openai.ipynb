# Run GPT-4 on diaries and quotes
We use `gpt-4` model to evaluate quotes for diaries (if quote is suitable for diary or not), and then train model on this data.

```python
import pandas as pd

df = pd.read_csv("../data/diaries_quotes_emb_twitter.csv")
```

```python
from tqdm import tqdm
from openai import OpenAI
client = OpenAI()

interactions = []

for i, e in tqdm(df.iterrows(), total=df.shape[0]):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "You are an expert recommendation evaluator. Now you must evaluate recommended quote of famous person for the person diary note. Just write '1' if quote is suitable, and '-1' if unsuitable"},
            {"role": "user", "content": f'Text: {e.Text}\n\nQuote: {e.Quote}'}
        ],
        max_tokens=2,
    )
    interactions.append(response.choices[0].message.content)

```

Add interactions to dataframe

```python
df519 = df[:len(interactions)]
df519['Interaction'] = interactions
```

Print interactions statistics

```python
df519["Interaction"].value_counts()
```
